## Regular Expression
ถ้าเป็น pattern อ.จะใช้ << >>

```python
re.search(<<patern>>,string)
re.findall(<<patern>>,string)
```
พื้นฐาน regex
```
^ เริ่มต้น
$ สิ้นสุด
. ตัวอะไรก็ได้
? มีตัวก่อนหน้าหรือไม่มีีก็ได้ เช่น <<^e-?mail$>>
+ มีหรือไม่มีก็ได้ (ตัวเดียว)
* มีหรือไม่มีก็ได้ (หลายตัว)
```

````python
import nltk,re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]
word_ed = [w for w in wordlist if re.search('ed$',w)]
print(word_ed)
````

````python
wlist = ['email','e-mail','emails','e-mails','electornic-mail','electornic mails','electonic-mails','chickenmail','mail','chickenmails','mailsd']
email = [w for w in wlist if re.search('^(?:|e.*)mail(?:|s)$',w)]
print(email)
````

````python
#โคตรเจ๋ง
# reg กดมือถือ T9 ปุ่ม 4 6 5 3 <<^[ghl][mno][jlk][def]$>>
word_T9_4 = [w for w in wordlist if re.search('^[ghl][mno][jlk][def]$',w)]
print(word_T9_4)
````

````python
finger_twisters = [w for w in wordlist if re.search('^[ghlmnojlk]+$',w)] #กดปุ่ม 4 5 6 แบบใดก็ได้
print(finger_twisters)
````

````python
chat_words = sorted(set(w for w in nltk.corpus.nps_chat.words()))
mine = [w for w in chat_words if re.search('^m+i+n+e+$',w)]
print(mine) #ตัวอย่างการลากเสียง
````

````python
#อยากได้ ha อย่างเดียว
ha = [w for w in chat_words if re.search('^h[ha]+$',w)]
print(ha)
````

````python
wsj = ['1','11','111','1111','11111']
print([w for w in wsj if re.search('^[0-9]{4}$',w)])#เลขยาว 4 ตัว
print([w for w in wsj if re.search('^[0-9]{2,4}$',w)])#เลขยาว 2-4 ตัว
print([w for w in wsj if re.search('^[0-9]{4,}$',w)])#เลขยาว 4 ตัวขึ้นไป
````

## Useful Application or Regular Expressions
### Extracting word pieces


````python
word = 'supercalifragilisticexpialidocious'
vowels= re.findall(r'[aeiou]',word)
print(len(vowels))
print(vowels)
````

````python
wsj = sorted(set(nltk.corpus.treebank.words()))
fd = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{3,{}}',word)) # บัคง่ะ lol
````

````python
print([int(n) for n in re.findall(r'[0-9]{2,4}','2009-12-32')])
````

````python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
csv = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]',w)]
cfd = nltk.ConditionalFreqDist(csv)
cfd.tabulate()
````

````python
# ตัด suffix
def stem(word):
    for suffix in ['ing','ly','ed','ious','ies','ive','es','s']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word
print(stem('processing'))
````

````python
def stem(word):
    result = re.findall(r'^(.*?)(ing|ly|ed|ious|ies|ive|es|s|ment)$',word)
    if len(result):
        return result[0][0]
    return word
````
