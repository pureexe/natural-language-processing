```python
#Exam 4
from nltk.corpus import inaugural,stopwords
def vocab(text):
    stopword = stopwords.words('english')
    return len(set([word.lower() for word in text if word.isalpha() and word not in stopword]))
print(vocab(inaugural.words()))
```

```python
#Exam9
from nltk import FreqDist,bigrams
from nltk.corpus import inaugural
def my_collocation(word,text):
    word_pair = list(bigrams([w for w in text if w.isalpha()]))
    interest_pair = [w for w in word_pair if w[0] == word]
    freq = FreqDist(interest_pair)
    most_pair = freq.most_common(1)[0]
    return word+" "+most_pair[0][1]+" "+str(most_pair[1])
print(my_collocation('America',inaugural.words()))
print(my_collocation('never',inaugural.words()))
```
