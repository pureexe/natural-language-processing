```python
## โจทย์ ให้สร้างฟังก์ชัน
# 1. lexcical diversity
def lexical_diversity(data):
    return len(data) / len(set(data))
# 2. หาสัดส่วนคำคิดเป็นเปอร์เซ็น
def percentage_of_word(word,data):
    return data.count(word) / len(data) * 100
```

```python
## สร้าง Frequency Distribution แบบอาจารย์
## ทำให้ string ตัวเล็ก "str".lower()
fdist1 = FreqDist(text1)
# แสดง 50 คำแรก
fdist1.most_common(50) 
# พลอต 50 คำแรก
fdist1.plot(50,cumulative=False)
```

```python
## Fine-Grained Selection of words
# ตัวอย่างหาคำที่ยาว 15 ตัวอักษรขึ้นไป
longword = [w for w in set(text1) if len(w) > 7 and fdist1[w] > 7]
print(sorted(longword))
```
