### Accessing Text from the Web and from Disk

```python
import nltk, re, pprint
from nltk import word_tokenize
from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"

response = request.urlopen(url)
raw = response.read().decode('utf8')
print(type(raw))
print(len(raw))
print(raw[:365])
```

```python
# ตัดคำ (Eng)
tokens = word_tokenize(raw)
print(type(tokens))
print(len(tokens))
print(tokens[:100])
```

```python
#หาตำแหน่งหัว
print(raw.find("PART I"))
#หาตำแหน่งท้าย
print(raw.rfind("End of Project Gutenberg"))
```

```python
# เรียกกระบวนการ timing
content =raw[5336:1157810]
print(content)
```

```python
# การ clean html (nltk.clean_html())
url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
html = request.urlopen(url).read().decode('utf8')
html[:500]
```

```python
#Beautify แม่มเลย (งี้ก็ได้หรอ) 
from bs4 import BeautifulSoup
DOM = BeautifulSoup(html,'html.parser')
raw = DOM.get_text()
```

```python

content = DOM.find_all('div',class_='bodytext')[0].get_text()
```

```python
print(content)
```

```python
content = DOM.find_all('div',class_='bodytext')[0].get_text()
content = nltk.Text(word_tokenize(content)) #อย่าลืม tokenize ก่อนนะจ๊ะ
```

```python
content.concordance("gene")
```

```python
content.collocations()
```

```python
## จากไฟล์ก็ได้
path = nltk.data.find('corpora/gutenberg/melville-moby_dick.txt')
f = open(path,'r') #ตัวอย่างใช้ rU แต่ depecate 
for line in f:
    print(line.strip())
    
```

```python
## หาคำศัพท์ (Vocab) จากข่าวเมื่อกี่
vocab = [word for word in content.vocab()]
print(vocab)
```
