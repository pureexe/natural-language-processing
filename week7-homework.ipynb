15.◑ Write programs to process the Brown Corpus and find answers to the following questions:

15.1 Which nouns are more common in their plural form, rather than their singular form? (Only consider regular plurals, formed with the -s suffix.)

```python
from nltk.corpus import brown
from nltk import FreqDist

noun_list = [word.lower() for word,tag in brown.tagged_words(tagset='universal') if tag == "NOUN"]
noun_plurals = [word for word in noun_list if word.endswith("s")]
noun_singula =  [word for word in noun_list if not word.endswith("s")]
freq_plurals = FreqDist(noun_plurals)
freq_singular = FreqDist(noun_singula)
result = [word for word in freq_plurals.keys() if freq_plurals[word] > freq_singular[word[:-1]]]
print(result)
#answer display below
```

15.2 Which word has the greatest number of distinct tags. What are they, and what do they represent?

```python
from nltk import ConditionalFreqDist
from nltk.corpus import brown

freq2D = ConditionalFreqDist(brown.tagged_words(tagset='universal'))
sortedDist = sorted([(name,len(freq2D[name])) for name in freq2D.keys()], key=lambda x: x[1],reverse=True)
print(sortedDist[0][0])
print(freq2D[sortedDist[0][0]].keys())

#answer below
#word has the greatest number of distinct tags: that
#What are they: ['ADP', 'PRON', 'DET', 'ADV', 'X']
#what do they represent: that have many difference meaning
```

15.4 Which tags are nouns most commonly found after? What do these tags represent?


```python
from nltk.corpus import brown
from nltk import FreqDist,bigrams

brown_news_tagged = brown.tagged_words(tagset = 'universal')
word_tag_pairs = list(bigrams(brown_news_tagged))
noun_preceders = [b[1] for (a,b) in word_tag_pairs if a[1] == 'NOUN'] 
fdist = FreqDist(noun_preceders)
print(fdist.most_common(1)[0])
#Answer below
#most commonly found: .
#What do these tags represent: i don't known, "." isn't describe in upenn tagset (https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html)
```

18.◑ Generate some statistics for tagged data to answer the following questions:

18.1 What proportion of word types are always assigned the same part-of-speech tag

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist
brown_news_tagged = brown.tagged_words(tagset = 'universal')
confreq = ConditionalFreqDist([(word.lower(),tag)for word,tag in brown_news_tagged])
single_tag = len([1 for word in confreq.conditions() if len(confreq[word]) == 1])
set_len = len(set([word.lower() for word,tag in brown_news_tagged]))
print(single_tag/set_len)
#Answer below
```

18.2 How many words are ambiguous, in the sense that they appear with at least two tags?

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist
freq2D = ConditionalFreqDist(brown.tagged_words(tagset='universal'))
ambiguous = [(name,len(freq2D[name])) for name in freq2D.keys() if len(freq2D[name]) >= 2]
print("ambiguous", len(ambiguous))
#Answer below
```

18.3 What percentage of word tokens in the Brown Corpus involve these ambiguous words?


```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist
wordlist = brown.tagged_words(tagset='universal')
freq2D = ConditionalFreqDist(wordlist)
ambiguous_set = set([name for name in freq2D.keys() if len(freq2D[name]) >= 2])
tokenlist =  [1 for word,tag in wordlist if word in ambiguous_set]
print(len(tokenlist)/len(wordlist)*100,'%')
#Answer below
```

20.Write code to search the Brown Corpus for particular words and phrases according to tags, to answer the following questions:

20.1 Produce an alphabetically sorted list of the distinct words tagged as MD.

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist

cfd = ConditionalFreqDist(brown.tagged_words())
md_words = [word for word in cfd.conditions() if cfd[word]['MD'] != 0]
print(md_words)
```

20.2 Identify words that can be plural nouns or third person singular verbs (e.g. deals, flies).

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist

cfd = ConditionalFreqDist(brown.tagged_words())
two_words = [word for word in cfd.conditions() if cfd[word]['NNS'] and cfd[word]['VBZ']]
print(two_words)
```

20.3 Identify three-word prepositional phrases of the form IN + DET + NN (eg. in the lab).

```python
from nltk.corpus import brown
from nltk import ConditionalFreqDist,trigrams

trigrams = list(trigrams(brown.tagged_words()))
for trigram in trigrams:
    zipped_tag = [t for t in zip(*trigram)]
    if zipped_tag[1] == ('IN', 'DT', 'NN'):
        print(zipped_tag[0])
```

20.4 What is the ratio of masculine to feminine pronouns?

```python
from nltk.corpus import brown
from nltk import FreqDist
fd = FreqDist(brown.words())

print((fd['he'] + fd['He']) / (fd['she'] + fd['She']))
```
